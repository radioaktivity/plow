```python
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib as mpl
dpi = 400
mpl.rcParams['figure.dpi']= dpi

font = {'family' : 'normal',
           'size'   : 2}

rc('font', **font)
```

```python

x = np.linspace(0,1,10)
y = np.linspace(0,1,10)

points = np.zeros([len(x)*len(y), 2])
k = 0
for i in x:
    for j in y:
        points[k,1] = j
        points[k,0] = i
        k += 1
```

```python
tri = Delaunay(points)


# This gives for triangle i the number of points 
# in the point array which build their borders
i = 69 # 25 26 36
print("corners of i", tri.simplices[i])

# This gives numbers of the neighboring triangles of triangle i
i = 69 # 140 71 68
print("neigbors of i", tri.neighbors[i])
```

```python

plt.triplot(points[:,0], points[:,1], tri.simplices)

plt.plot(points[:,0], points[:,1], 'o')
for j, p in enumerate(points):

    plt.text(p[0], p[1], j, ha='right') # label the points

for j, s in enumerate(tri.simplices):

    p = points[s].mean(axis=0)

    plt.text(p[0], p[1], '#%d' % j, ha='center') # label triangles

plt.xlim(-0.5, 1.5); plt.ylim(-0.5, 1.5)
mpl.rc("savefig", dpi=dpi)
```

```python
def get_tangent(x,y,dxdy, width=0.1):
    x_tan = np.linspace(x-width,x+width,10)
    print(x_tan)
    y_tan = dxdy*(x_tan - np.ones([1,10])[0]*x) + np.ones([1, 10])[0]*y
    print(y_tan)
    return [x_tan, y_tan]

plt.scatter([1,2], [1,0.5])
tan1 = get_tangent(1,1,-2)
tan2 = get_tangent(2,0.5,1)
plt.plot(tan1[0],tan1[1])
plt.plot(tan2[0],tan2[1])
```
