```python
from scipy.interpolate import griddata
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
j = 20
grid_x, grid_y = np.meshgrid(np.linspace(0,100,300), np.linspace(0,100,300))

n = 50
x = np.linspace(0,100,n)
y = np.linspace(0,100,n)
X, Y = np.meshgrid(x,y)
points = np.vstack([X.ravel(), Y.ravel()]).T

rng = np.random.default_rng()

points = rng.random((n*n, 2))*100

values = []
for i in range(n*n):
    values.append(random.uniform(0,10))
```

```python


grid_z2 = griddata(points, values, (grid_x, grid_y), method='nearest')
```

```python

plt.imshow(grid_z2.T, origin='lower', cmap=plt.cm.BuPu_r)
```
